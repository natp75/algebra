```python
#homework_3_3 Напишите уравнение плоскости, параллельной данной и проходящей через начало координат.
```

```python
Ax+By+Cz+D=0
В этом случае координаты нормального вектора плоскости будут соответствовать коэффициентам перед x, y, z.
Так как плоскости параллельны, то любой нормальный вектор первой плоскости будет соответствовать нормальному вектору второй плоскости.
Примем данный вектор как вектор нормали второй плоскости.
Тогда уравнение плоскости, проходящей через точку   и с нормальным вектором  примет вид:
A(x-x0)+B(y-y0)+C(z-z0)+D=0


Пусть задана плоскость: A1x + B1y + C1z + D1 = 0
и прямая: (x-x1)/(x2-x1)=(y-y1)/(y2-y1)=(z-z1)/(z2-z1). Как узнать, принадлежит прямая плоскости или нет?
нужно решить систему  уравнений.
если система имеет единственное решение, значит прямая  пересекает плоскость  в одной точке.
если система не имеет решения, то значит прямая  параллельна плоскости .
если система имеет множество решений, то значит прямая  принадлежит плоскости .
```

```python
%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt
from pylab import *
from mpl_toolkits.mplot3d import Axes3D
fig = figure()
ax = Axes3D(fig)
X = np.arange(-40, 40, 2)
Y = np.arange(-40, 40, 2)
X, Y = np.meshgrid(X, Y)
Z2 = 4 - 2*X - 3*Y
Z1 = -2*(X+10) - (Y+20) * 3
ax.plot_surface(X, Y, Z1)
ax.plot_surface(X, Y, Z2)
show()
```
