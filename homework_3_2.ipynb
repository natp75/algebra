```python
Доказательство. Пусть точки M1(x1, y1) и M2(x2, y2) посредством ортогонального преобразования
(𝑋=𝑎11 𝑥 + 𝑎12 𝑦+𝑎13 и 𝑌=𝑎21 𝑥 + 𝑎22 𝑦+𝑎23) переводятся соответствено в точки M'1 (x'1, y'1) и M'2(x'2, y'2). 
Требуется доказать, что отрезки M1M2 и M'1M'2 имеют равные длины. 
С помощью формул (𝑋=𝑎11 𝑥 + 𝑎12 𝑦+𝑎13 и 𝑌=𝑎21 𝑥 + 𝑎22 𝑦+𝑎23) и (a211 + a221 = 1, a212 + a222 = 1, a11a12 + a21a22 = 0)
получаем 
|M1M2|^2=[x2-x1]^2+[y2-y1]^2+[z2-z1]^2+[a11(x2-x1)+a12(y2-y1)]^2+[a21(x2-x1)+a22(y2-y1)]^2=(x2-x1)^2+(y2-y1)^2=|M1M2|^2
```
